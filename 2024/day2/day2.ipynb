```python
f = open("input_day2.txt", 'r')
data = f.readlines()
#data[:5]
```

```python
data2=[]
for s in data:
    s = s.strip('\n').split()
    s2 = [int(i) for i in s]
    data2.append(s2)
#data2[:5]

#listona_di_liste = [list(map(int, sublist)) for sublist in listona_di_liste]
```

```python
def monotona(l):
    if l == sorted(l) or l == sorted(l, reverse=True):
        return True
```

```python
def numeri_ok(l):
    return all([1 <= abs(l[i] - l[i + 1]) <= 3 for i in range(len(l)-1)])
```

```python
cont = 0
for lista in data2:
    if monotona(lista) and numeri_ok(lista):
        cont += 1
    else:
        for i in range(len(lista)):
            lista1 = lista.copy()
            lista1.pop(i)
            if monotona(lista1) and numeri_ok(lista1):
                cont += 1
                break

cont
```
