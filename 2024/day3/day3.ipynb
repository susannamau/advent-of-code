# Part 1

```python
import re
```

```python
f = open("input_day3.txt", 'r')
data = f.readlines()
#data
```

```python
s = "".join(data)
```

```python
# pattern = r"mul\((\d{1,3}),(\d{1,3})\)"
# matches = re.findall(pattern, s)
# result = sum(int(a)*int(b) for a,b in l)
```

```python
l = re.findall("mul\([0-9]{1,3},[0-9]{1,3}\)", s)
#l
```

```python
l = [x.strip('mul(') for x in [x.strip(')') for x in l]]
l
```

```python
somma = 0
for x in l:
    a,b = x.split(',')
    somma += int(a)*int(b)
```

```python
somma
```

# Part 2

```python
pattern_do = "do\(\)"
pattern_dont = "don't\(\)"
pattern_mul = "mul\([0-9]{1,3},[0-9]{1,3}\)"
```

```python
l = re.findall(pattern_mul+"|"+pattern_do+"|"+pattern_dont, s)
```

```python
active = True
somma = 0

for el in l:
    if el == "do()":
        active = True
        #print(">>>>>attivato<<<<<<")
    elif el == "don't()":
        active = False
        #print(">>>>>disattivato<<<<<<")
    else:
        #print(active)
        if active:
            a,b = el.strip('mul(').strip(')').split(',')
            #print(a,b)
            somma += int(a)*int(b)
somma
```
